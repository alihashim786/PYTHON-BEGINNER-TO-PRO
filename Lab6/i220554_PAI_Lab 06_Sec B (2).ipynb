### Kind
* 'line' (default): Line plot.
* 'bar': Vertical bar plot.
* 'barh': Horizontal bar plot.
* 'hist': Histogram.
* 'box': Box plot.
* 'kde': Kernel Density Estimation plot.
* 'density': Same as 'kde'.
* 'area': Area plot.
* 'scatter': Scatter plot.
* 'hexbin': Hexagonal bin plot.

### Marker
* 'o': Circle marker (default).
* 's': Square marker.
* '^': Triangle marker.
* '*': Star marker.
* '+': Plus marker.
* 'x': X marker.
* 'D': Diamond marker.
* '.': Point marker.

```python
img_scaled = img * 255

# Display the scaled image
plt.imshow(img_scaled.astype('uint8'))

# Add a colorbar for reference (optional)
plt.colorbar()

# Add title
plt.title('Image Example (Scaled to [0, 255])')

# Show the image
plt.show()
```

## Histogram

```python
# Convert the image to grayscale (assuming it's a color image)
img_gray = np.mean(img_scaled, axis=2)

# Flatten the 2D image array to a 1D array
img_flat = img_gray.flatten()

# Create a histogram
plt.hist(img_flat, bins=256, range=(0, 255), color='b')

# Add labels and title
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of the Image')

# Show the histogram
plt.show()
```

### Histogram Equilization

```python
# Load an image
image = cv2.imread('w3.png', cv2.IMREAD_GRAYSCALE)

# Plot the original and equalized images side by side
plt.figure(figsize=(10, 5))
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.show()
```

```python
# Load an image
image = cv2.imread('w3.png', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Plot the original and equalized images side by side
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

```python

# Create a figure with subplots for the images and histograms
fig, axes = plt.subplots(1, 2, figsize=(15, 5))


# Plot the histogram of the original image
axes[0].hist(image.ravel(), bins=256, range=(0, 256), color='b')
axes[0].set_xlabel('Pixel Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram of Original Image')

# Plot the histogram of the equalized image
axes[1].hist(equalized_image.ravel(), bins=256, range=(0, 256), color='b')
axes[1].set_xlabel('Pixel Value')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of Equalized Image')

plt.tight_layout()
plt.show()
```

## 3D Charts

* Read image again in RGB this time. Take first row of 2nd chanel and 5th row of third chanel and draw a line chart between them. Use at least two different kind and marker

```python
# YOUR CODE HERE



# Load the image in RGB mode
image = Image.open('ali.jpg').convert('RGB')
pixels = np.array(image)

# Extract the first row of the 2nd channel (G) and the 5th row of the 3rd channel (B)
green_channel_row = pixels[0, :, 1]
blue_channel_row = pixels[4, :, 2]

# Create a line chart
plt.figure(figsize=(10, 5))

# Line chart with a solid line and blue color
plt.plot(green_channel_row, label='Green Channel (Row 1)', color='blue', linestyle='-', marker='o')

# Line chart with a dashed line and red color
plt.plot(blue_channel_row, label='Blue Channel (Row 5)', color='red', linestyle='--', marker='x')

# Add labels and legend
plt.xlabel('Pixel Index')
plt.ylabel('Pixel Value')
plt.title('Line Chart of Green and Blue Channels')
plt.legend()

plt.show()
```
