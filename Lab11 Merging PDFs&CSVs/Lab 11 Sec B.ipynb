```python
def replace_whitespace (data):
    #replace \n and \t with a space
    string=""
    for page in data:
        string +=" "+ page.replace ("\n", " ").replace ("\t", " ")

    return string
```

```python

import PyPDF2

pdfFileObj = open('data.pdf', 'rb')
pdfReader = PyPDF2.PdfReader(pdfFileObj)
pages = pdfReader.pages
pages_text = [(pages[i].extract_text()) for i in range(len(pages))]
pages_texts = replace_whitespace(pages_text)
pages_texts

```
