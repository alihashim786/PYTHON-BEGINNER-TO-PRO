## Tokenization
### Sentence Tokenization
Sentence tokenization is the process of splitting text into sentences.

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

text = "NLTK is a powerful library. It helps with text processing."
sentences = sent_tokenize(text)
print(sentences)
```

### Word Tokenization
Word tokenization is the process of splitting sentences into words.

```python
from nltk.tokenize import word_tokenize

sentence = "NLTK is a powerful library."
words = word_tokenize(sentence)
print(words)
```

## Stopwords Removal
Stopwords are common words (e.g., "the," "and," "is") that are often removed from text.

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

print(stop_words)
```

```python
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

## Stemming and Lemmatization
Stemming reduces words to their root form, while lemmatization reduces them to their base or dictionary form.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

word = "describing"
stemmed_word = stemmer.stem(word)
lemmatized_word = lemmatizer.lemmatize(word, pos="v")

print(f"Stemmed: {stemmed_word}, Lemmatized: {lemmatized_word}")
```

## ngrams
ngrams gives you the freuency of not only a single word but more than one words as specified

```python
text = """The Young King
The Birthday of the Infanta
The Fisherman and his Soul
The Star-child"""
```

```python
import nltk
nltk.download('punkt')
```

```python
# normalizung the data
text = text.lower()
text
```

```python
# character wise frequency distribution
myFD = nltk.FreqDist(text)
myFD
```

```python
# generating tokens before creating ngrams
tokens = nltk.word_tokenize(text)
tokens
```

```python
# unigrams
unigrams = nltk.ngrams(tokens, 1)
unigramFD = nltk.FreqDist(unigrams)

for token in list(unigramFD.items()):
    print(token[0], token[1])
```

```python
# bigrams
bigrams = nltk.ngrams(tokens, 2)
bigramsFD = nltk.FreqDist(bigrams)

for bigram in list(bigramsFD.items()):
    print(bigram[0], bigram[1])
```

```python
# trigrams
trigrams = nltk.ngrams(tokens, 3)
trigramsFD = nltk.FreqDist(trigrams)

for trigram in list(trigramsFD.items()):
    print(trigram[0], trigram[1])
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.probability import FreqDist

# Ensure that 'text' is a string
history = history.lower()

# Find sentences containing 'Kerak'
kerak_sentences = [sentence for sentence in sent_tokenize(text) if 'Kerak' in sentence]

# Tokenize each sentence into unigrams and convert to lowercase for case insensitivity
unigrams = [word_tokenize(sentence.lower()) for sentence in kerak_sentences]

# Flatten the list of unigrams
unigrams = [word for sentence in unigrams for word in sentence]

# Create a FreqDist to count unigram occurrences
freq_dist = FreqDist(unigrams)

# Filter and count unigrams in sentences containing 'Kerak'
kerak_unigrams = [word for word in unigrams if any(word in sentence for sentence in kerak_sentences)]

# Get the three lowest occurring unigrams
lowest_occurrence_unigrams = freq_dist.most_common()[-3:]

print(f"Three lowest occurring unigrams in sentences containing 'Kerak':")
for unigram, count in lowest_occurrence_unigrams:
    print(f"{unigram}: {count}")
```

```python
# Tokenize the text into sentences
sentences = sent_tokenize(history)

filtered_tokens = []

# DefinING the stopwords set
stop_words = set(stopwords.words('english'))

# InitializING a flag to track consecutive sentences with "Saladin" or "king"
consecutive_flag = 0

for sentence in sentences:
    # TokenizING the sentence into words
    words = word_tokenize(sentence)

    # CheckING if the sentence contains 'Saladin' or 'king'
    if 'Saladin' in words or 'king' in words:
        consecutive_flag = consecutive_flag + 1
    else:
        consecutive_flag = 0

    # If two consecutive sentences contain 'Saladin' or 'king', removing  stopwords in those sentences
    if consecutive_flag == 2:

        filtered_words = [word for word in words if word.lower() not in stop_words]
        filtered_tokens.extend(filtered_words)

    if consecutive_flag < 2:
        # If not consecutive, add the words as they are
        filtered_tokens.extend(words)

# Print the filtered tokens
print(filtered_tokens)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Tokenize the text into words
words = word_tokenize(history)

# Initialize the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]

# Initialize counters for words that remain the same
same_words_stemming = 0
same_words_lemmatization = 0

# Perform stemming and lemmatization
stemmed_words = [stemmer.stem(word) for word in filtered_words]
lemmatized_words = [lemmatizer.lemmatize(word, pos='a') for word in filtered_words]

# Count words that remain the same after stemming
for original, stemmed in zip(filtered_words, stemmed_words):
    if original == stemmed:
        same_words_stemming += 1

# Count words that remain the same after lemmatization
for original, lemmatized in zip(filtered_words, lemmatized_words):
    if original == lemmatized:
        same_words_lemmatization += 1

# Print the results
print(f"Number of words that remain the same after stemming: {same_words_stemming}")
print(f"Number of words that remain the same after lemmatization: {same_words_lemmatization}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

geopolitical_aspects = ["Jerusalem", "king", "queen", "valley", "Saladin"]

# Create a dictionary to store the frequency of each aspect
aspect_frequency = {aspect: filtered_words.count(aspect) for aspect in geopolitical_aspects}

# Convert the dictionary to a DataFrame for better visualization
aspect_df = pd.DataFrame.from_dict(aspect_frequency, orient='index', columns=['Frequency'])

# Sort the DataFrame by frequency
aspect_df = aspect_df.sort_values(by='Frequency', ascending=False)

# Plot the frequencies using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(aspect_df.index, aspect_df['Frequency'], color='skyblue')
plt.xlabel('Geopolitical Aspects')
plt.ylabel('Frequency')
plt.title('Frequency of Geopolitical Aspects in the Text')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
